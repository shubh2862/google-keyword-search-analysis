```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import cv2

from pytrends.request import TrendReq
trends = TrendReq()

%matplotlib inline
```

```python
trends.build_payload(kw_list=["cillian murphy"])
data = trends.interest_by_region()
data = data.sort_values(by="cillian murphy", ascending=False)
data = data.head(30)
print(data)
```

```python
data.reset_index().plot(x="geoName", y="cillian murphy",
                        figsize=(20,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.rc('font', size=16)
plt.title('Total Google Searches for Pornhub', fontweight='bold')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['cillian murphy'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 15))
data['cillian murphy'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Pornhub', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
